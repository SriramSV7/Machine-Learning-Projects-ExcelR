```python
plt.figure(figsize=(8, 8))
bu_counts = df['BU'].value_counts().sort_values(ascending=False)
sns.barplot(x=bu_counts.index, y=bu_counts.values, palette='viridis')
plt.ylim(0, 250)
plt.title('Frequency Distribution of Business Unit (BU)')
plt.xlabel('Business Unit (BU)')
plt.ylabel('Count of Transactions')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
```

```python
plt.figure(figsize=(10, 6))
brand_counts = df['Brand'].value_counts().sort_values(ascending=False)
sns.barplot(x=brand_counts.index,y=brand_counts, palette='magma')
plt.title('Frequency Distribution of Transactions by Brand')
plt.xlabel('Brand')
plt.ylabel('Count of Transactions')
plt.tight_layout()
```
