```python
import numpy as np
from scipy import stats
```

```python
# 1. Given Data

mu_theoretical = 1000 + (5 * 600) 
sigma = 5 * 25                    
n = 25                            
sample_mean = 3050               
alpha = 0.05
```

```python
# 2. Calculate the Test Statistic (z)
# Formula: z = (sample_mean - mu) / (sigma / sqrt(n))
standard_error = sigma / np.sqrt(n)
z_score = (sample_mean - mu_theoretical) / standard_error
```

```python
# 3. Determine Critical Value and P-value
z_critical = stats.norm.ppf(1 - alpha)  # Right-tailed test
p_value = 1 - stats.norm.cdf(z_score)
```

```python
# 4. Output Results
print(f"Theoretical Mean (mu): {mu_theoretical}")
print(f"Test Statistic (z): {z_score:.4f}")
print(f"Critical Value (z_critical): {z_critical:.4f}")
print(f"P-value: {p_value:.4f}")
```

```python
# 5. Decision and Conclusion
if z_score > z_critical:
    print("\nDecision: Reject the Null Hypothesis (H0)")
    print("Conclusion: There is strong evidence that weekly operating costs are higher.")
else:
    print("\nDecision: Fail to Reject the Null Hypothesis (H0)")
    print("Conclusion: There is no strong evidence to support the claim that costs are higher.")
```
