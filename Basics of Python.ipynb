```python
a=int(input("enter the number"))
if a==2:
    print(a,"is a prime number")
elif a<=1:
    print(a,"is not a prime number")
else:
    
    for i in range(2,int(a**0.5)+1):
        if a%i==0:
            print(a,"is not a prime")
            break
    else:
        print(a,"is a prime number")
        
```

```python
room_charges=2000
doc_con_fee=1500
lab_charges=300
medicine=2400

a=int(input("enter how many days you stay:"))
a*=room_charges
b=int(input("how many test:"))
b*=lab_charges
bill_before_discount=a+doc_con_fee+b+medicine
print("your room cost is:",a,"per day is:",room_charges)
print("your doctor consultation charge is:",doc_con_fee)
print("your lab charge is:",b,"charge per test:",lab_charges)
print("your medicine cost is: ",medicine)
if bill_before_discount>10000:
    c=bill_before_discount/10
    final_bill=bill_before_discount-c
    print("before discount:",bill_before_discount)
    print("Final bill is:",final_bill)
else:
    print("Final bill is:",bill_before_discount)
```
