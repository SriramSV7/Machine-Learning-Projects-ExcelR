```python
## Given:

import numpy as np
import scipy.stats as stats

data = np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])
```

```python
# Basic statistics

n = len(data)
mean = np.mean(data)
std_dev = np.std(data, ddof=1)
```

```python
confidence_level = 0.99
df = n - 1
t_critical = stats.t.ppf((1 + confidence_level) / 2, df)
margin_of_error_t = t_critical * (std_dev / np.sqrt(n))
ci_t = (mean - margin_of_error_t, mean + margin_of_error_t)

print(f"\na)")
print(f"t-critical value: {t_critical:.3f}")
print(f"Margin of Error: {margin_of_error_t:.4f}")
print(f"Confidence Interval: {ci_t}")
```

```python
pop_std_dev = 0.2
z_critical = stats.norm.ppf((1 + confidence_level) / 2)
margin_of_error_z = z_critical * (pop_std_dev / np.sqrt(n))
ci_z = (mean - margin_of_error_z, mean + margin_of_error_z)
```

```python
print(f"\nb)")
print(f"z-critical value: {z_critical:.3f}")
print(f"Margin of Error: {margin_of_error_z:.4f}")
print(f"Confidence Interval: {ci_z}")
```
