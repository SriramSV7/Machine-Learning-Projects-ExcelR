### Dimensionality Reduction with PCA

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
feature = df.drop(columns="Type")
target = df["Type"]
```

```python
### scaling

scaler = StandardScaler()

feature_scaled = scaler.fit_transform(feature)
```

```python
# 2. Implement PCA

pca = PCA()
pca_transformed = pca.fit_transform(feature_scaled)
```

```python
# 3. Determine optimal components (Scree plot / Cumulative variance)

explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)

plt.figure(figsize=(10, 6))
plt.bar(range(1, len(explained_variance) + 1), explained_variance, alpha=0.5, align='center', label='Individual explained variance')
plt.step(range(1, len(cumulative_variance) + 1), cumulative_variance, where='mid', label='Cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal component index')
plt.title('Scree Plot & Cumulative Explained Variance')
plt.legend(loc='best')
```

```python
# Transform original dataset into principal components (Top 3 as commonly used for this dataset)
pca_3 = PCA(n_components=3)
X_pca = pca_3.fit_transform(feature_scaled)
pca_df = pd.DataFrame(X_pca, columns=['PC1', 'PC2', 'PC3'])
pca_df['Type'] = target.values
pca_df.to_csv('pca_transformed_data.csv', index=False)
```

```python
print("Explained variance ratio (first 3):", pca_3.explained_variance_ratio_)
print("Total variance explained (first 3):", sum(pca_3.explained_variance_ratio_))
```

### Clustering with Original Data

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
wcss_original = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(feature_scaled)
    wcss_original.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss_original)
plt.title('Elbow Method (Original Data)')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

```python
# Apply K-means (k=3)

kmeans_orig = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=42)
orig_clusters = kmeans_orig.fit_predict(feature_scaled)
```

```python
# Evaluation metrics for Original Data

sil_orig = silhouette_score(feature_scaled, orig_clusters)
sil_orig
```

### Clustering with PCA Data

```python
# Determine optimal k using Elbow Method for PCA data
wcss_pca = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(X_pca)
    wcss_pca.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss_pca)
plt.title('Elbow Method (PCA Data)')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

```python
# Apply K-means (k=3) to PCA data

kmeans_pca = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=42)
pca_clusters = kmeans_pca.fit_predict(X_pca)
```

```python
# Evaluation metrics for PCA Data
sil_pca = silhouette_score(X_pca, pca_clusters)
sil_pca
```

```python
### Visualization: PCA Clusters (using PC1 and PC2)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=orig_clusters, cmap='viridis', edgecolors='k')
plt.title('K-means Clustering (Original Data Scaled)')
plt.xlabel('PC1')
plt.ylabel('PC2')

plt.subplot(1, 2, 2)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=pca_clusters, cmap='viridis', edgecolors='k')
plt.title('K-means Clustering (PCA Data)')
plt.xlabel('PC1')
plt.ylabel('PC2')

plt.tight_layout()
```
